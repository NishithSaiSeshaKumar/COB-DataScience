Removing the Outliers Using **Z-scores**

Setting a threshold of 3 standard deviations from the mean, and filetring out the outliers in the **release_year** 

```python
numerical_columns = 'release_year'
mean = data[numerical_columns].mean()
std = data[numerical_columns].std()
threshold = 3 * std
data = data[abs(data[numerical_columns] - mean) <= threshold]
```
