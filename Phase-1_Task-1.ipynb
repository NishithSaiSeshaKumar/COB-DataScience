#**Task 1**: Call this public Api and create a csv dataset using python and pandas
Api: - https://data.binance.com/api/v3/ticker/24hr

Install The required packages:
<br> 
*Requests*: To make HTTPS requests to Api<br>
*Pandas*: To create a Dataframe with the obtained Data

**Importing packages**

```python
import requests
import pandas as pd
```

Placing the url to the API in the variable **api_url**

```python
api_url = "https://data.binance.com/api/v3/ticker/24hr"
response = requests.get(api_url)
```

Checking Whether the API is working by making a request to the API

```python
if response.status_code == 200:
    data = response.json()
else:
    print("Failed to retrieve data from the API")
    data = []
```

Viewing the Data retrieved from the API

```python
response.content
```

Creating a Dataframe **df** with the Data retrieved from the API

```python
df = pd.DataFrame(data)
```

Viewing the Data in DataFrame

```python
df.head()
```

```python
df.columns
```

Saving the Dataframe as a CSV file

```python
csv_filename = 'api_data.csv'
df.to_csv(csv_filename, index=False)
```
